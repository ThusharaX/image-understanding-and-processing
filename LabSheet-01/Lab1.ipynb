### Question 1 - read and save the image from/in storage device.

```python
# import the necessary packages
import cv2
import matplotlib.pyplot as plt
```

```python
# Read the image
img = cv2.imread("ring.jpg", 1) # 1 for color image
```

```python
# Prewiew the image - Method 1
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB)) # Convert BGR to RGB
```

```python
# Prewiew the image (separate window) - Method 2
cv2.imshow("Ring Window", img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Save the image
cv2.imwrite("ring_copy.jpg", img)
```

### Question 2 - access pixel values of an image and modify them.

```python
# Read the image
img_blue = cv2.imread("blue.jpg", 1) # 1 for color image

# access pixel values of an image
pix = img_blue[10, 10]

print(pix) # prints the BGR value of the pixel
```

```python
# change the pixel values
img_blue[10, 10] = [0, 0, 255] # Red

# Prewiew the image
plt.imshow(cv2.cvtColor(img_blue, cv2.COLOR_BGR2RGB)) # Convert BGR to RGB
```

### Question 4 - split and merge channels

```python
# Split channels
b, g, r = cv2.split(img)
```

```python
# threashold value for blue color
_, bmod = cv2.threshold(b, 100, 255, cv2.THRESH_BINARY)
```

```python
# Merge channels back
merged = cv2.merge((bmod, g, r))
```

```python
# preview the image
plt.imshow(cv2.cvtColor(merged, cv2.COLOR_BGR2RGB))
```

### Question 5 - resize/rotate image.

```python
# Resize the image
resized = cv2.resize(img, (50, 50))  # width, height

# preview the image
plt.imshow(cv2.cvtColor(resized, cv2.COLOR_BGR2RGB))
```

```python
# Rotate the image by 90 degrees and clockwise
rotated = cv2.rotate(img, cv2.ROTATE_90_CLOCKWISE)  # COUNTERCLOCKWISE other one

# preview the image
plt.imshow(cv2.cvtColor(rotated, cv2.COLOR_BGR2RGB))
```

### Question 6 -  draw circle/rectangle/lines.

```python
# drawing circle --> img name,potision center, radius, BGR color, thickness of line
cv2.circle(img, (300, 250), 80, (255, 0, 0), 4)

# drawing rectangle --> img name, top-left corner, bottom-right corner, BGR color, thickness of line
cv2.rectangle(img, (100, 100), (200, 200), (0, 255, 0), 3)

# drawing line --> img name, start point, end point, BGR color, thickness of line
cv2.line(img, (150, 150), (200, 200), (0, 0, 255), 3)

# preview the image
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
```

### Question 7 - write text on image.

```python
# Read the image
img = cv2.imread("ring.jpg", 1)

# Write text on image --> img name, text, starting point, font name, font size, color
cv2.putText(img, "Sample Text", (100, 100), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 0, 255))

# preview the image
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
```

### Question 8 - covert color to grayscale/binary/negative image

```python
# Turn to gray scale --> 1 = BGR, -1 = Original, 0 = Gray
img_gray = cv2.imread("ring.jpg", 0)

# preview the image
plt.imshow(cv2.cvtColor(img_gray, cv2.COLOR_BGR2RGB))
```

```python
# Turn into a binary image
_, binaryimg = cv2.threshold(img_gray, 100, 255, cv2.THRESH_BINARY)

# preview the image
plt.imshow(cv2.cvtColor(binaryimg, cv2.COLOR_BGR2RGB))
```

```python
# Turn into a negative image
negative = cv2.bitwise_not(img)

# preview the image
plt.imshow(cv2.cvtColor(negative, cv2.COLOR_BGR2RGB))
```
